This notebook is used to display images in one of the predicted cluster

```python
import json
import os
import pandas as pd
import ipyplot
from tools import url_to_image_id_map_helper
```

```python
url_to_image_id_map = url_to_image_id_map_helper()
```

First, set the `expr` to the experiment name:
- dog3_non_random_with_init
- dog3_non_random_from_scratch
- dog3_random_with_init
- birds

```python
# experiment name
expr = "birds"
```

```python
with open(os.path.join("results", expr + "_" + "tracker.json")) as tracker_handle:
    tracker = json.load(tracker_handle)
```

```python
print("There are {} clusters.".format(len(tracker["clusters"])))
```

```python
predicted_cluster_id = 2
```

```python
urls_in_cluster = [("." if "Birds" in data["value"] else "") + data["value"]
                   for data in tracker["clusters"][predicted_cluster_id]]
labels = [url_to_image_id_map[url] for url in urls_in_cluster]
```

```python
print("This cluster has %d images" % len(urls_in_cluster))
```

```python
ipyplot.plot_images(urls_in_cluster,
                    labels,
                    max_images=len(urls_in_cluster),
                    img_width=150)
```
